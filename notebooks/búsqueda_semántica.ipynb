```python
#Abrir documento con pandas
import pandas as pd

df = pd.read_csv('/content/df_total.csv')
print(df.head())

document = df["news"].tolist()
print (len (document))
documents = document[:]
```

```python
#Instalar torch y sentence tranformer
!pip uninstall -y torch torchvision torchaudio

!pip install torch torchvision torchaudio --index-url https://download.pytorch.org/whl/cpu
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("paraphrase-multilingual-MiniLM-L12-v2",
                            device="cpu")

embeddings = model.encode(
    documents,
    batch_size=64,
    show_progress_bar=True
)
```

```python
#Igualar y agrupar vectores con FAISS
import faiss
import numpy as np

dimension = embeddings.shape[1]

index = faiss.IndexFlatL2(dimension)

index.add(np.array(embeddings))
```

```python
#Función para la búsqueda semántica
def search(query, k=5):

    query_embedding = model.encode([query])

    distances, indices = index.search(np.array(query_embedding), k)

    results = []

    for i in indices[0]:
        results.append(document[i])

    return results

search ("odio", 5)
```
